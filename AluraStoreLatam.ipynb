### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
ventas_por_cat = tienda.groupby('Categoría del Producto')['Producto'].count()

print(ventas_por_cat.sort_values(ascending=False))
```

```python
ventas_ordenadas = ventas_por_cat.sort_values(ascending=False)

plt.figure(figsize=(10 ,6))
ventas_ordenadas.plot(kind='bar', color='skyblue')
plt.title('Cantidad de productos vendidos por categoria')
plt.xlabel('Categoria del Producto')
plt.ylabel('Cantidad de Productos Vendidos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
ventas_por_cat.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Distribución de Categorias Vendidas')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ventas_por_cat.index, ventas_por_cat.values, marker='o', linestyle='-', color='darkgreen')
plt.title('Comparativa de Ventas por Categoría')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Productos Vendidos')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Muebles es la categoría más vendida en todas las tiendas, superando consistentemente las 440 unidades. Tienda 3 lidera con 499 ventas, lo cual sugiere una fuerte demanda constante por este tipo de producto y una buena capacidad de surtido.

Electrónicos ocupa el segundo lugar en ventas en todas las tiendas. Tiendas 3 y 4 empatan en 451 unidades, lo que indica una competencia directa y alta rotación de productos tecnológicos. Es una categoría clave para promociones y actualizaciones de inventario.

Juguetes y Electrodomésticos se disputan el tercer y cuarto lugar en todas las tiendas.

Tienda 4 destaca en Juguetes con 338 unidades, posiblemente por una mejor estrategia estacional o de marketing en ese rubro.

Electrodomésticos tiene una leve caída en Tienda 4 (254 unidades), lo que podría indicar menor stock, menor demanda o precios poco competitivos.

Deportes y diversión se mantiene estable entre 275 y 284 unidades en todas las tiendas. Aunque no lidera, presenta una demanda constante y podría escalar con campañas dirigidas o combos promocionales.

Instrumentos musicales presenta mayor variabilidad.

Tienda 2 tiene un pico con 224 unidades, mientras que Tienda 4 baja a 170.
Esto puede reflejar una diferencia en enfoque de público objetivo o disponibilidad de inventario.

Libros y Artículos para el hogar se ubican en los últimos lugares en ventas.

Tienda 2 destaca en Libros (197) y Tienda 4 en Artículos para el hogar (201).
Esto sugiere que podrían ser segmentos de oportunidad para fortalecer o especializarse según la demanda local.

Recomendaciones Estratégicas
Tienda 3 se perfila como la más sólida en las principales categorías: Muebles, Electrónicos y Juguetes.

Tienda 2 podría impulsar sus categorías menos fuertes (Artículos para el hogar y Electrodomésticos) con promociones o alianzas con proveedores.

Tienda 4 destaca en Juguetes y Artículos para el hogar, por lo que puede especializarse en público familiar o infantil.

Todas las tiendas deberían evaluar oportunidades en Instrumentos musicales, ya que hay una diferencia significativa en ventas y podría haber mercado sin explotar.
